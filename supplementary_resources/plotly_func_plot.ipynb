```python
colors = ['#2F2C49', '#77C4BE', '#C96F4B', '#6F656E', '#8B9F93', '#E5CB6A', '#2B4FA3', '#000000']
```

```python
def save_func_to_file(label_f, x_min, x_max, out_file_path, num_samples=1000):
    fig = plt.figure(figsize=(16, 9))
    plt.grid(True, color='#000', alpha=0.2)
    ax = plt.gca()
    ax.set_xlim(x_min, x_max)
    
    x = np.linspace(x_min, x_max, num_samples)
    for label, (func, color) in label_f.items():
        y = func(x)
        line, = plt.plot(
            x, y,
            linewidth=2,
            label=label,
            color=color
        )
#         ax.fill_between(x, y, 0, facecolor='#77C4BE', alpha=0.5, interpolate=True)
        ax.legend()
    plt.savefig(fname=out_file_path, format='svg')
    plt.show()
```

```python
def original_func(x):
    return 1/(2 * np.pi) + np.cos(x - np.pi) ** 3 / (3 * np.pi) + (x - np.pi) ** 2 / (9 * np.pi) * np.sin(x - np.pi) ** 3
```

```python
def cdf(x):
    return -1/(9 * np.pi) * (
        (-0.75 * x ** 2 + 4.71 * x - 5.9) * np.cos(x) + \
        (0.083 * x ** 2 - 0.52 * x + 0.8) * np.cos(3 * x) - \
        4.5 * x + 1.5 * x * np.sin(x) - 0.056 * x * \
        np.sin(3 * x) - 2.46 * np.sin(x) + 0.42 * np.sin(3 * x)
    )
```
